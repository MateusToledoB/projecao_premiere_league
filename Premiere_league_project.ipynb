Abrindo os arquvios .csv

```python
import pandas as pd

table_partidas_jogadas = pd.read_excel('/content/premiere_confrontos.xlsx')
table_partidas_ajogar = pd.read_excel('/content/partidas_a_jogar.xlsx')
```

Removendo valores nulos da tabela de jogos pendentes

```python
table_partidas_ajogar.fillna('a jogar', inplace=True)
```

mudando os valores das colunas com o NÃºmero de gols para variaveis do tipo inteiro (INT)

```python
display(table_partidas_jogadas)
```

```python
table_partidas_jogadas['Gols casa'] = table_partidas_jogadas['Gols casa'].astype(int)
table_partidas_jogadas['gols fora'] = table_partidas_jogadas['gols fora'].astype(int)
display(table_partidas_jogadas)
```

Calculando as mÃ©dias das partidas feitas em casa.

```python
#aqui realizei um agrupamento dos valores presentes na coluna 'casa' calculando tambÃ©m a mÃ©dia das colunas que possuem valores numericos, ignorando assim a coluna 'fora'
#o resultado gera um datatset com os indices sendo os nomes das equipes e as duas colunas contendo as mÃ©dias feitas dentro de casa.
media_gols_casa = table_partidas_jogadas.groupby('casa').mean(numeric_only=True)
display(media_gols_casa)
```

```python
#Renomeei as colunas do dataset 'media_gols_casa' utilizando um dicionÃ¡rio para a relaÃ§Ã£o nome antigo x novo nome, facilitando a leitura.

media_gols_casa = media_gols_casa.rename(columns={'Gols casa' : 'gols feitos em casa', 'gols fora' : 'gols sofridos em casa'})
```

```python
display(media_gols_casa)
```

Calculando media das partidas feitas fora

```python
#mesmo passo realizado anteriormente, porÃ©m o agrupamento foi feito sob a coluna  'fora'.
media_gols_fora = table_partidas_jogadas.groupby('fora').mean(numeric_only=True)
```

```python
display(media_gols_fora)
```

```python
media_gols_fora = media_gols_fora.rename(columns={'Gols casa':'gols sofridos fora', 'gols fora':'gols feitos fora'})
```

Mesclando as informaÃ§Ãµes

```python
#Uma parte importante do projeto, a junÃ§Ã£o de todas as mÃ©dias de cada equipe.
#para isso criei um novo dataset mesclando as tabelas 'media_gols_casa' e 'media_gols_fora' utilizando os indices de ambas (no caso o nome das equipes) as tabelas para realizar a junÃ§Ã£o.
table_estatisticas = media_gols_casa.merge(media_gols_fora, left_index=True, right_index=True)

display(table_estatisticas)
```

```python
#aqui utilizei o comando .reset-index para transformar os valores agrupados em uma coluna novamente.
table_estatisticas = table_estatisticas.reset_index()
#Renomeei a coluna criada acima para facilitar a leitura.
table_estatisticas = table_estatisticas.rename(columns={'casa': 'time'})
display(table_estatisticas)
```

DistribuiÃ§Ã£o de poison: calculando a probabilidade dos resultados com lambda

```python
#InstalaÃ§Ã£o da biblioteca scipy, uma biblioteca usada em Python para computaÃ§Ã£o cientÃ­fica e anÃ¡lise de dados.
#A biblioteca scipy complementa a biblioteca numpy e fornece funcionalidades adicionais para otimizaÃ§Ã£o, Ã¡lgebra linear, estatÃ­sticas, processamento de sinais, entre outros.
!pip install scipy
```

```python
# utilizando a implementaÃ§Ã£o da distribuiÃ§Ã£o de Poisson da biblioteca SciPy.
#obtendo as mÃ©dias de cada time podemos usar a distribuiÃ§Ã£o de poisson para calcular
#a probabilidade de diferentes resultados.
from scipy.stats import poisson
```

```python
#criando uma funÃ§Ã£o para agir sobre as linhas do dataset
def calcular_pontuaÃ§Ã£o_esperada(linha):
#atribuimos Ã  variÃ¡vel time_casa o valor presente na coluna 'casa' para a linha especÃ­fica passada como argumento.
  time_casa = linha['casa']
#atribuimos Ã  variÃ¡vel time_fora o valor presente na coluna 'fora' para a linha especÃ­fica passada como argumento.
  time_fora = linha['fora']



#aqui vamos calcular o lambda de ambos os times que vÃ£o se enfrentar.
#o lambda calcula qual Ã© a mÃ©dia de gols que o time x jogando em casa faz contra o time y jogando fora de casa.
#podemos calcular esse valor multiplicando a mÃ©dia de gols feitos pelo time x em casa com a mÃ©dia de gols sofridos pelo time y fora de casa(e vise versa).

#atribuimos o valor de lambda do time da casa localizando a mÃ©dia de gols feitos em casa usando o comando .loc[] para filtrar [linha,coluna] do dataset.
#localizamos a mÃ©dia de gols sofridos pelo time de fora com o mesmo comando (.loc[l])
  lambda_casa = table_estatisticas.loc[table_estatisticas['time']== time_casa, 'gols feitos em casa'].iloc[0]*table_estatisticas.loc[table_estatisticas['time']== time_fora, 'gols sofridos fora'].iloc[0]

  lambda_fora = table_estatisticas.loc[table_estatisticas['time']== time_fora, 'gols feitos fora'].iloc[0]*table_estatisticas.loc[table_estatisticas['time']== time_casa, 'gols sofridos em casa'].iloc[0]



  pv_casa = 0
  p_empate = 0
  pv_fora = 0
#utilizaÃ§Ã£o de dois loops for para inteirar sobre diferentes valores de gols marcados pelo time da casa e o time de fora
#ambos os valores variam de 0 - 7, representando os valores possiveis de 0 a 7 gols para cada time.
  for gols_casa in range (8):
    for gols_fora in range(8):

      probabilidade_resultado = poisson.pmf(gols_casa, lambda_casa)* poisson.pmf(gols_fora, lambda_fora)
    #aqui calculamos a probabilidade do resultado multiplicando o numero de gols de cada time pelo seu lambda(valor esperado daquele time fazer x gols naquele jogo em especifico)

    #aqui somamos a porcentagem de cada placar acontecer de acordo com as variaveis criadas acima (pv_casa, p_empate e pv_fora)
    #caso o nÃºmero de gols gerados para ambos os times forem iguais, a porcentagem deste resultado ira ser acresentada na variavel de empate, caso o nÃºmero de gols do time da casa
    #for maior que o nÃºmero de gols gerados pelo time de fora, a porcenatgem de chance deste resultado sera acrescentada na variavel de vitÃ³ria do time da casa, o mesmo vale caso
    #o nÃºmero de gols do time de fora se sobresaia sobre o nÃºmero de gols do time da casa.

      if gols_casa == gols_fora:
        p_empate += probabilidade_resultado
      elif gols_casa > gols_fora:
        pv_casa += probabilidade_resultado
      elif gols_casa < gols_fora:
        pv_fora += probabilidade_resultado


#calculo do valor esperado de pontos do time da casa e do time de fora.
#este calculo Ã© realizado multiplicando a probabilidade de vitÃ³ria de cada time por 3 (valor que uma equipe recebe caso ganhe uma partida) + a probabilidade do empate(caso a equipe empate ela tambÃ©m recebe pontos)
  ve_casa = pv_casa * 3 + p_empate

  ve_fora = pv_fora * 3 + p_empate


#agora criamos novas colunas a serem adicionadas no dataset que vamos usar de parametro para essa funÃ§Ã£o, as novas colunas vÃ£o armazenar o valor esperado
#de pontos dos times de casa e de fora.
  linha['pontos_casa'] = ve_casa
  linha['pontos_fora'] = ve_fora
  return linha


```

```python
#aqui foi aplicada a funÃ§Ã£o criada acima no datset que contÃ©m os jogos pendentes da cmapeonado, calculando assim o valor esperado de cada equipe nos jogoso que ainda irÃ£o realizar.
#axis=1 indica que a funÃ§Ã£o deve ser rodada sob o eixo 1 do dataset, ou seja, a cada linha.
table_partidas_ajogar = table_partidas_ajogar.apply(calcular_pontuaÃ§Ã£o_esperada, axis=1)

display(table_partidas_ajogar)
```

```python
#abertura do arquivo que contÃ©m a tabela de classificaÃ§Ã£o do campeonato.
table_classificacao = pd.read_excel('/content/premiere_table.xlsx')
display(table_classificacao)
```

```python
#criaÃ§Ã£o de uma nova tabela de classificaÃ§Ã£o contendo apenas o nome das equipes e seus pontos
table_classificacao_att =table_classificacao[['Equipe', 'Pts']]
display(table_classificacao_att)
```

```python
#criaÃ§Ã£o de duas tabelas, contendo o numero de pontos feitos em casa e pontos feitos fora de casa
#agrupando os valores da coluna 'casa' da tabela de jogos pendentes somando os valores numericos da coluna 'pontos_casa'
pontuacao_casa = table_partidas_ajogar.groupby('casa').sum(numeric_only = True)[['pontos_casa']]
display(pontuacao_casa)
#agrupamento dos valores presentes na coluna 'fora' somando os vallres numericos da coluna 'pontos' fora
pontuacao_fora = table_partidas_ajogar.groupby('fora').sum(numeric_only=True)[['pontos_fora']]
display(pontuacao_fora)
```

```python
#criaÃ§Ã£o de uma nova funÃ§Ã£o para atualizar o numero de pontos de cada time
def atualizar_pontuacao(linha):

  time = linha['Equipe']

  #a soma da pontuaÃ§Ã£o vai ser calculada somando os pontos jÃ¡ existentes com a pontuaÃ§Ã£o feita jogando em casa + a pontuaÃ§Ã£o feita fora de casa
  #para localizar a pontuaÃ§Ã£o de cada time vamos armazenar dentro da coluna time os valores presentes na coluna 'equipe' do dataset que vai ser usado como paraemtro da funÃ§Ã£o
  #a pontuaÃ§Ã£o feita em casa e fora de casa do time que corresponde ao mesmo nome do equipe do dataset Ã© somada ao total.

  pontuacao = linha['Pts'] + float(pontuacao_casa.loc[time, 'pontos_casa']) + float(pontuacao_fora.loc[time, 'pontos_fora'])
  return pontuacao

#aqui atualizamos a coluna de pontos aplicando a funÃ§Ã£o de soma criada acima.
table_classificacao_att['Pts'] = table_classificacao_att.apply(atualizar_pontuacao, axis=1)
display(table_classificacao_att)



```

```python
#realizada uma nova ordenaÃ§Ã£o na tabela de classificaÃ§Ã£o, classificando/ordenando os valores da coluna 'Pts' na ordem decrescente (ascentende = falso)
table_classificacao_att = table_classificacao_att.sort_values(by='Pts', ascending=False)

#atualizando os valores do Ã­ndices para uma melhor visualizaÃ§Ã£o, somando 1 a cada Ã­ndice.
table_classificacao_att = table_classificacao_att.index +1
display(table_classificacao_att)
```
